```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

emissions_by_sector = pd.read_csv('co2_emissions_sector.csv')
emissions_by_fossil = pd.read_csv('co2_emissions_fossils.csv')
electricity = pd.read_csv('Germany_Electricity.csv')
```

```python
emissions_by_sector.head()
```

```python
#We can see that there are some missing values

emissions_by_sector.info()
```

```python
emissions_by_sector.loc[emissions_by_sector['Year'] <= 2000, 'Fugitive Emissions'] = emissions_by_sector['Fugitive Emissions'].fillna(emissions_by_sector.loc[emissions_by_sector['Year'] <= 2000,'Fugitive Emissions'].mean())
```

```python
emissions_by_sector.loc[((emissions_by_sector['Year'] > 2000) & (emissions_by_sector['Year'] <= 2010)), 'Fugitive Emissions'] = emissions_by_sector['Fugitive Emissions'].fillna(emissions_by_sector.loc[((emissions_by_sector['Year'] > 2000) & (emissions_by_sector['Year'] <= 2010)),'Fugitive Emissions'].mean())
```

```python
emissions_by_sector.loc[((emissions_by_sector['Year'] > 2010) & (emissions_by_sector['Year'] <= 2020)), 'Fugitive Emissions'] = emissions_by_sector['Fugitive Emissions'].fillna(emissions_by_sector.loc[((emissions_by_sector['Year'] > 2010) & (emissions_by_sector['Year'] <= 2020)),'Fugitive Emissions'].mean())
```
